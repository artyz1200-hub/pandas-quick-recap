The IQR method is the standard way of identifying outliers.

we take the ‘normal range’ of data, and anything that falls significantly outside it is discarded.

````python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

print("IQR:", IQR)

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR


outliers = df[(df['Fare'] < lower) | (df['Fare'] > upper)] 

print(f'IQR range: [{lower:.1f}, {upper:.1f}]')
print(f'Outliers Fare: {len(outliers)} rowas ({len(outliers)/len(df)*100:.1f}%)')
print(f'Max Fare in data: {df["Fare"].max():.0f}')
````

Later we can decide what to do with it,maybe leave,maybe replace with upper

````python
print("Survival rate by Pclass:")

print(df.groupby('Pclass')['Survived'].mean().round(2))
print()
````

````python
df.Age.describe()
````

````python
print("Survival rate by age group:")
df.groupby(pd.cut(df["Age"], bins=[0,12,18,35,60,100]))['Survived'].mean().round(2)
````

````python
print('Age and Fare by sex:')

print(df.groupby('Sex_label')[['Age', 'Fare']].mean().round(2))
````

````python
# --- agg: gives us the aggregated values ---


result = df.groupby('Pclass').agg(
    count=('Passengerid', 'count'),
    survival_rate=('Survived', 'mean'),
    avg_age=('Age', 'mean'),
    avg_fare=('Fare', 'mean'),
    max_fare=('Fare', 'max'),
    min_fare=('Fare', 'min')
).round(2)


display(result)
````

````python
Sex_label = df.groupby("Sex").agg(
    count=('Passengerid', 'count'),
    survival_rate=('Survived', 'mean'),
    avg_age = ('Age', 'mean'),
    avg_fare = ('Fare', 'mean'),
    min_age = ('Age', 'min'),
    max_age = ('Age', 'max')

)

display(Sex_label)
````

````python
Embarked_check = df.groupby("Embarked").agg(
    count=('Passengerid', 'count'),
    survival_rate=('Survived', 'mean'),
    avg_fare = ('Fare', 'mean'),
    mean_fare = ('Fare', 'mean'),

    max_class = ('Pclass', lambda x: x.mode()[0])
).round(2)

display(Embarked_check)
````

````python
# we can also group by multiple columns and apply different aggregation functions to each column. For example, we can group by both 'Pclass' and

result2 = (
    df.groupby(['Pclass', 'Sex_label'])['Survived']
    .agg(['mean', 'count'])
    .rename(columns={'mean': 'survival_rate', 'count': 'num_passengers'})
    .round(3)
    .sort_values('survival_rate', ascending=False)
    .reset_index()
)

display(result2)
````

## 6. merge / join/

````python
passengers = df[['Passengerid', 'Age', 'Survived', 'Pclass']].head(10).copy()

class_info = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'Description': ['Upper class', 'Middle class', 'Lower class'],
    "avg_ticket_price": [100, 50, 25]
})

#inner join - only rowa with matching Pclass values in both DataFrames

merged = passengers.merge(class_info, on='Pclass', how='inner')
print('Inner join:', merged.shape)
display(merged.head())
````

````python
# LEFT JOIN — all rows from the left DataFrame (passengers) are included in the result, and matching rows from the right DataFrame (class_info) are included where available. If there is no match, the result will contain NaN for the columns from the right DataFrame.
cabin_data = pd.DataFrame({
    'Passengerid': [1, 2, 3],
    'cabin': ['C85', 'E46', None]
})

merged_left = passengers.merge(cabin_data, on='Passengerid', how='left')
print('Left join (all 10 passengers):', merged_left.shape)
display(merged_left)
````

````python
df_a = pd.DataFrame({'id': [1, 2, 3],           'val_a': ['x', 'y', 'z']})
df_b = pd.DataFrame({'passenger_id': [1, 2, 4], 'val_b': ['a', 'b', 'c']})

merged_keys = df_a.merge(df_b, left_on='id', right_on='passenger_id', how='outer')
display(merged_keys)



print('\nTypes of join: inner | left | right | outer (analogues SQL JOIN)')
````

## 7. pivot_table

`groupby` would produce a list of rows, whereas `pivot_table` produces a cross-tabulation table , which is easier to read when you need to compare two variables at the same time.

````python
pt = pd.pivot_table(
    df,
    values  = 'Survived',
    index   = 'Pclass',
    columns = 'Sex_label',
    aggfunc = 'mean',
    margins = True,   # строка/колонка итогов 'All'
).round(3)

print('Survival rate по Pclass x Sex:')
display(pt)
````

````python
pt2 = pd.pivot_table(
    df,
    values  = ['Age', 'Fare'],
    index   = 'Pclass',
    columns = 'Sex_label',
    aggfunc = {'Age': 'mean', 'Fare': 'median'},
).round(2)

display(pt2)


pt2_flat = pt2.stack(level=1).reset_index()
print('\n After stack + reset_index:')
display(pt2_flat.head())
````

## 8.Map, Apply, Assign

`map` replaces each value in a column according to a dictionary.

````python
df['Pclass_name'] = df['Pclass'].map({1: 'First', 2: 'Second', 3: 'Third'})
print(df[['Pclass', 'Pclass_name']].value_counts().sort_index())
````

`apply`  applies the function to each element/string

````python
df['FamilySize_fast'] = df['SibSp'] + df['Parch'] + 1

# np.where - if/else
df['IsAlone'] = np.where(df['FamilySize_fast'] == 1, 1, 0)

# pd.cut - cut to bins with labels
df['Fare_bin'] = pd.cut(
    df['Fare'],
    bins=[0, 10, 30, 100, 1000],
    labels=['low', 'mid', 'high', 'vip']
)

# pd.qcut - cut to quantiles
df['Age_quartile'] = pd.qcut(df['Age'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

display(df[['Fare', 'Fare_bin', 'Age', 'Age_quartile', 'IsAlone']].head(8))
````

`assign` , adds new columns without modifying the original `df`, and returns a new one.
Within `assign`, `x` refers to the current DataFrame at this stage of the pipeline. Each column is computed using the lambda expression `x`

````python
df_features = (
    df[['Passengerid', 'Age', 'Fare', 'SibSp', 'Parch', 'Sex', 'Pclass', 'Survived']]
    .assign(
        FamilySize   = lambda x: x['SibSp'] + x['Parch'] + 1,
        IsAlone      = lambda x: (x['SibSp'] + x['Parch'] == 0).astype(int),
        FarePerPerson= lambda x: (x['Fare'] / x['FamilySize']).round(2),
        AgeGroup     = lambda x: pd.cut(
            x['Age'], bins=[0, 18, 60, 100], labels=['child', 'adult', 'senior']
        ),
    )
    .dropna(subset=['AgeGroup'])
    .reset_index(drop=True)
)

display(df_features.head())
print('Shape:', df_features.shape)
````

````python

result = (
    df
    .query('Age > 0')                           #filter
    .assign(
        FamilySize = lambda x: x['SibSp'] + x['Parch'] + 1,
        IsAlone    = lambda x: (x['SibSp'] + x['Parch'] == 0).astype(int),
    )
    .groupby(['Pclass', 'Sex_label'])
    .agg(
        n              = ('Passengerid', 'count'),
        survival_rate  = ('Survived', 'mean'),
        pct_alone      = ('IsAlone', 'mean'),
        avg_family     = ('FamilySize', 'mean'),
    )
    .round(3)
    .sort_values('survival_rate', ascending=False)
    .reset_index()
)

display(result)
````
